cleaning the time taken column

```python
dataset["Time_taken(min)"] = dataset["Time_taken(min)"].str.replace("(", "")
dataset["Time_taken(min)"] = dataset["Time_taken(min)"].str.replace("min", "")
dataset["Time_taken(min)"] = dataset["Time_taken(min)"].str.replace(")", "")
dataset["Time_taken(min)"]= dataset["Time_taken(min)"].astype("Int64")
```

Filling missing value

```python
mean=dataset["Delivery_person_Age"].mean()
mean=np.int64(mean)
#dataset['Delivery_person_Age']=dataset['Delivery_person_Age'].replace("<NA>",0,inplace=True)
dataset['Delivery_person_Age'].fillna(0,inplace=True)

dataset['Delivery_person_Age']=dataset['Delivery_person_Age'].replace(0,mean)
```

```python
#changing the data type of Delivery_person_Ratings from obj to int
dataset["Delivery_person_Ratings"] = dataset["Delivery_person_Ratings"].astype("float")
#finding mean of "Delivery_person_Ratings"
mean=dataset["Delivery_person_Ratings"].mean()

mean
```

```python
#filing the Delivery_person_Ratings empty column
dataset["Delivery_person_Ratings"]=dataset["Delivery_person_Ratings"].replace(np.nan,mean)
```

```python
null_count_new=dataset.isnull().sum()
null_count_new
```

```python
#replaving missing value with the mod of Wetherconditions in the column
mode_value = dataset['Weatherconditions'].mode().iloc[0]
mode_value

# Replace NaN values with the mode
dataset['Weatherconditions'].fillna(mode_value, inplace=True)
```

```python
#replaving missing value with the mod of Road_traffic density in the column
mode_value = dataset['Road_traffic_density'].mode().iloc[0]
mode_value

# Replace NaN values with the mode
dataset['Road_traffic_density'].fillna(mode_value, inplace=True)
```

```python
#finding mean in multiple deliveries column
mode=dataset["multiple_deliveries"].mode().iloc[0]
mode
#filling dataset with the mode value
dataset["multiple_deliveries"].fillna(mode,inplace=True)
```

```python
#finding mean in Festivsal column
mode=dataset["Festival"].mode().iloc[0]
mode
#filling dataset with the mode value
dataset["Festival"].fillna(mode,inplace=True)
```

```python
#filling dataset with the mode value
dataset["City"].fillna(mode,inplace=True)
```
